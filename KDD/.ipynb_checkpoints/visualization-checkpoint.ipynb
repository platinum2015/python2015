## MULTIPLOT

```python
import matplotlib.pyplot as plt

#FIGURE e' la cornice
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

#IT ALWAYS refer to the LAST ONE!!!!
#AX3
plt.plot(randn(50).cumsum(), 'k--')
#AX1
_ = ax1.hist(randn(100), bins=20, color='k', alpha=0.3)
#AX2
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
```

## BAR PLOT

```python
#Bar plot using Pandas obj method
import pandas as pd
fig, axes = plt.subplots(2, 1)
data = pd.Series(floor(np.random.rand(16)*10), index=list('abcdefghijklmnop')) 

print(data.values)
data.plot(kind='bar', ax=axes[0], color='g', alpha=0.7) 
data.plot(kind='barh', ax=axes[1], color='k', alpha=0.7)
```

## HISTOGRAM

```python
# using Pandas obj method
comp1 = np.random.normal(0, 1, size=200) # N(0, 1) 

comp2 = np.random.normal(10, 2, size=200) # N(10, 4) 

values = pd.Series(np.concatenate([comp1, comp2])) 

values.hist(bins=100, alpha=0.3, color='b', normed=True) 

values.plot(kind='kde', style='k--') 
```

## SCATTER

```python
# using Pandas obj method
#tsmodels project, select a few variables, then compute log differences: 
macro = pd.read_csv('H:/Technology/python/Dataset/pydata-book-master/ch08/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']] 
trans_data = np.log(data).diff().dropna() 
trans_data[-5:]# Out[89]: 
#cpi 198 -0.007904 199 -0.021979 200 0.002340 201 0.008419 202 0.008894 
#m1 tbilrate 0.045361 -0.396881 0.066753 -2.277267 0.010286 0.606136 0.037461 -0.200671 0.012202 -0.405465 
#unemp 0.105361 0.139762 0.160343 0.127339 0.042560 
plt.scatter(trans_data['m1'], trans_data['unemp']) 
#Out[91]: <matplotlib.collections.PathCollection at 0x43c31d0>
plt.title('Changes in log %s vs. log %s' % ('m1', 'unemp')) 


df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
               columns=['A', 'B', 'C', 'D'],
               index=np.arange(0, 100, 10)) 
df.plot() 
```

## FILES

```python

#SAVE to File
plt.savefig('figpath.png', dpi=400, bbox_inches='tight')
#WRITE to file (example dynamic web page)
from io import StringIO buffer = StringIO() plt.savefig(buffer)
plot_data = buffer.getvalue() 




fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) 
data = pd.read_csv('ch08/spx.csv', index_col=0, parse_dates=True) spx = data['SPX'] 
spx.plot(ax=ax, style='k-') 
crisis_data = [
(datetime(2007, 10, 11), 'Peak of bull market'), (datetime(2008, 3, 12), 'Bear Stearns Fails'), (datetime(2008, 9, 15), 'Lehman Bankruptcy') 
] 
for date, label in crisis_data:
ax.annotate(label, xy=(date, spx.asof(date) + 50), 
xytext=(date, spx.asof(date) + 200), arrowprops=dict(facecolor='black'), horizontalalignment='left', verticalalignment='top') 
# Zoom in on 2007-2010 ax.set_xlim(['1/1/2007', '1/1/2011']) ax.set_ylim([600, 1800]) 
ax.set_title('Important dates in 2008-2009 financial crisis') 

```

# PIE CHART

```python
import matplotlib.pyplot as plt
labels = 'NO change win', 'change win'
fracs = [33,100]
plt.pie(fracs, labels=labels)
patches, texts, autotexts = plt.pie(fracs, labels=labels,
                                    autopct='%.0f%%',
                                    shadow=True)
autotexts[0].set_color('y')
patches[1].set_color('y')
plt.show()
```
