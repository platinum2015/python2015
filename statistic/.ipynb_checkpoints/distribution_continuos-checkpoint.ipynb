## LOGISTIC

```python
from scipy.stats import logistic
import matplotlib.pyplot as plt

# dir(logistic) #all  methods

mean, var, skew, kurt = logistic.stats( moments='mvsk')#mvsk specify all the 4
stddev=logistic.std()
#median=logistic.ppf()
median=1
#n percentile binom.ppf( qth ,n,p)

#Display the probability mass function (pDf):
print("LOGISTIC:  mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(stddev)+
" skew:" + str(skew) + " kurt:" + str(kurt)+ " median:" + str(median) )



x = np.arange(0,20)

#PDF
data = logistic.pdf(x)
f_pdf = plt.figure()
plt.title("PdF" )
plt.plot(x,data,'-o')

#CDF
mycdf=logistic.cdf( x)
f_cdf = plt.figure()
plt.title("CDF" )
plt.plot(mycdf)

plt.show()
```

# RANDOM WALK

```python
import random
#VERSION 1
position = 0
walk = [position]
steps = 1000
for i in xrange(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
plot(walk)
print("VERSION 1:position",position)

#VERSION 2
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
plot(walk)
print("VERSION 2:walk",walk.min(),walk.max())
```

## BETA

```python
#External: generate Randomly  or from File
from scipy.stats import beta
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)

a=2
b=0.5

mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
print("BETA:  mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(sqrt(var))+" skew:" + str(skew) + " kurt:" + str(kurt))

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
ax.plot(x, beta.pdf(x, a, b),'r-', lw=5, alpha=0.6, label='beta pdf')
```

```python
fig = plt.figure();
ax1  = fig.add_subplot(2, 2, 1) 
ax2  = fig.add_subplot(2, 2, 2) 
ax3  = fig.add_subplot(2, 2, 3) 
ax4  = fig.add_subplot(2, 2, 4) 

# a<1 & b<1
a=0.7
b=0.6
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
print("BETA(a<1 & b<1) : a="+str(a)+" b="+str(b)
      +" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(sqrt(var))+" skew:" + str(skew) + " kurt:" + str(kurt))

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
ax1.set_ylabel('a<1 & b<1') 
ax1.plot(x, beta.pdf(x, a, b))


# a=b
a=2
b=2
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
print("BETA(a=b):   a="+str(a)+" b="+str(b)
      +" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(sqrt(var))+" skew:" + str(skew) + " kurt:" + str(kurt))

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
ax2.set_ylabel('a=b') 
ax2.plot(x, beta.pdf(x, a, b))

# a>b
a=3
b=2
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
print("BETA (a>b):a="+str(a)+" b="+str(b)
      +" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(sqrt(var))+" skew:" + str(skew) + " kurt:" + str(kurt))

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
ax3.set_xlabel('a>b') 
ax3.plot(x, beta.pdf(x, a, b))

# a<b
a=2
b=300
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
print("BETA (a<b):  a="+str(a)+" b="+str(b)
      +" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(sqrt(var))+" skew:" + str(skew) + " kurt:" + str(kurt))

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
ax4.set_xlabel('a<b') 
ax4.plot(x, beta.pdf(x, a, b))
```

```python
#Calculated
from scipy.stats import beta
figs, axs = plt.subplots(1, 1)
a=2
b=300

r = beta.rvs(a, b, size=1000)
axs.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
axs.legend(loc='best', frameon=False)
plt.show()
```

## CHI SQUARE

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)

df = 55 #degree of freedom
mean, var, skew, kurt = chi2.stats(df, moments='mvsk')

print("CHI SQUARE "
      +" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(sqrt(var))+" skew:" + str(skew) + " kurt:" + str(kurt))

x = np.linspace(chi2.ppf(0.01, df),chi2.ppf(0.99, df), 100)
ax.plot(x, chi2.pdf(x, df),'r-', lw=5, alpha=0.6, label='chi2 pdf')
```

```python
fig, ax = plt.subplots(1, 1)
df = 55 #degree of freedom
r = chi2.rvs(df, size=1000)

ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

##uniform

```python
def uniform_cdf(x):
    "returns the probability that a uniform random variable is <= x"
    if x < 0: return 0 # uniform random is never less than 0
    elif x < 1: return x # e.g. P(X <= 0.4) = 0.4
    else: return 1 # uniform random is always less than 1
```
