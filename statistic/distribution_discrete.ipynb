## BINOMIAL

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# dir(binomial) #all  methods

n = 10
p = 0.5
#loc=200 add 200 to mean

mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')#mvsk specify all the 4
stddev=binom.std(n,p)
median=binom.ppf(0.5 ,n,p)
#n percentile binom.ppf( qth ,n,p)

#Display the probability mass function (pmf):
print("BINOMIAL: n="+str(n)+" p="+str(p)+" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(stddev)+
" skew:" + str(skew) + " kurt:" + str(kurt)+ " median:" + str(median) )

x = np.arange(0, n * 1.1)

#PMF
data = binom.pmf(x,n,p)
f_pmf = plt.figure()
plt.title("PMF" )
plt.plot(x,data,'-o')

#CDF
mycdf=binom.cdf( x,n, p)
f_cdf = plt.figure()
plt.title("CDF" )
plt.plot(mycdf)

plt.show()



```

## HYPERGEOMETRIC

```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
dir(hypergeometric)

w = 40
b = 60
k=  20

mean, var, skew, kurt = hypergeom.stats( w + b  , w ,k, moments='mvsk')#mvsk specify all the 4
stddev=hypergeom.std(w + b  , w ,k)
median=hypergeom.ppf(0.5 ,w + b  , w ,k)
#n percentile hypergeom.ppf(qth ,w + b  , w ,k)

var  = "%.2f" % var
mean = "%.2f" % mean
skew = "%.2f" % skew
kurt = "%.2f" % kurt
#Display the probability mass function (pmf):
print("HyperGeometric:  mean:"+str(mean) + " median:"+ str(median) +" var:"+ str(var) + " stddev:"+ str(stddev)+
" skew:" + str(skew) + " kurt:" + str(kurt) )



#PMF
rv=hypergeom( w + b  , w ,k)
x = np.arange(0,k+1)
data = rv.pmf(x)

f_pmf = plt.figure()
plt.title("Hypergeometric PMF  (w:"+str(w)+",b:"+str(b)+",k:"+str(k)+")")
plt.plot(x,data,'-o')

#CDF
mycdf=hypergeom.cdf(x,w + b  , w ,k)
f_cdf = plt.figure()
plt.title("CDF" )
plt.plot(mycdf)

plt.show()
```

## GEOMETRIC

```python
from scipy.stats import geom
import matplotlib.pyplot as plt
#dir(hypergeometric)

p = 0.67

mean, var, skew, kurt = geom.stats( p, moments='mvsk')#mvsk specify all the 4
stddev=geom.std(p)
median=geom.ppf(p,0.5)
#n percentile geom.ppf(qth ,w + b  , w ,k)

var  = "%.2f" % var
mean = "%.2f" % mean
skew = "%.2f" % skew
kurt = "%.2f" % kurt
#Display the probability mass function (pmf):
print("Geometric:  mean:"+str(mean) + " median:"+ str(median) +" var:"+ str(var) + " stddev:"+ str(stddev)+
" skew:" + str(skew) + " kurt:" + str(kurt) )

#PMF
rv=geom(p)
x = np.arange(0,p*10,1)
data = rv.pmf(x)

f_pmf = plt.figure()
plt.title("geometric PMF  (p:"+str(p)+")")
plt.plot(x,data,'-o')

#CDF
mycdf=geom.cdf(x,p)
f_cdf = plt.figure()
plt.title("CDF" )
plt.plot(mycdf)

plt.show()
```

## POISSON

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

# dir(poisson) #all  methods

lambd = 2
#loc=200 add 200 to mean

mean, var, skew, kurt = poisson.stats(lambd, moments='mvsk')#mvsk specify all the 4
stddev=poisson.std(lambd)
median=poisson.ppf(0.5 ,lambd)
#n percentile nbinom.ppf( qth ,n,p)

#Display the probability mass function (pmf):
print("POISSON: lambd="+str(lambd)+" mean:"+str(mean) + " var:"+ str(var) + " stddev:"+ str(stddev)+
" skew:" + str(skew) + " kurt:" + str(kurt)+ " median:" + str(median) +")" )


x = np.arange(0, lambd * 2)

#PMF
data = poisson.pmf(x,lambd)
f_pmf = plt.figure()
plt.title("PMF" )
plt.plot(x,data,'-o')

#CDF
mycdf=poisson.cdf( x,lambd)
f_cdf = plt.figure()
plt.title("CDF" )
plt.plot(mycdf)

plt.show()
```
